To give you a better idea to how the social platform works, here are some summary points and definitions:

The input parameter is the name of a text file containing all of Chirper’s profile information. You may assume the file is in the same directory as this notebook. The file is a series of profile data, where data for each profile has the following format:<br>
`USERID`<br>

`FOLLOWED1, FOLLOWED2, ..., FOLLOWEDm`<br>
where `USERID`, `FOLLOWERi`, and `FOLLOWEDj` are all unique numbers. The *third line* represents the users which *follow* `USERNAME`, and the *fourth line* represents the users which `USERNAME` *follows*. In the text file there is always a blank line between different profiles. Based off this data, the function constructs a `profiles_dictionary` where the orders of the ids representing followers and users follows, are the same orders as which they appear in the text file. See `profiles.txt`, which is included in the same directory as this notebook, as an example. You may assume that the given text file is perfect and that there are no formatting errors.

```python
# DONE, BUT DOESN'T PASS TESTS
def create_profile_dictionary(file_name):
    f = open(file_name, 'r')
    line = f.readlines()
    new_list = [l.replace('\n',"") for l in line]
    new_dict = {}

    i = 0
    while i < len(line)-3:
        x = i + 1
        y = i + 2
        z = i + 3
        if new_list[y] != '':
            y1 = [new_list[y]]
        else:
            y1 = []

        if new_list[z] != '':
            z1 = [new_list[z]]
        else:
            z1 = []

        new_dict[int(new_list[i])] = ((new_list[x],y1,z1))
        i += 5
    

    profiles_dictionary = {}
    for key, value in new_dict.items():
        name, list1, list2 = value
        list1 = [int(num) for num in list1[0].split(',')] if list1 else []
        list2 = [int(num) for num in list2[0].split(',')] if list2 else []
        profiles_dictionary[key] = (name, list1, list2)


    f.close()
    return(profiles_dictionary)

create_profile_dictionary('test1.txt')
```

#### Question 2: `create_chirp_dictionary(file_name) -> chirp_dictionary`
`create_chirp_dictionary(file_name)` opens the file `file_name` in the current working directory and reads the content into a `chirp dictionary`.<br>
The input parameter is the name of a text file containing all of Chirper’s message information. You may assume the file is in the same directory as this notebook. The file is a series of message data, where data for each message has the following format:<br>
`CHIRPID`<br>
`USERID`<br>
`MESSAGE`<br>
`TAG1, TAG2, ..., TAGk`<br>
`LIKED1, LIKED2, ..., LIKEDn`<br>
`DISLIKED1, DISLIKED2, ..., DISLIKEDm`<br>
where `CHIRPID`, `USERID`, `LIKEDi`, and `DISLIKEDj` are all unique numbers. `USERID` is the id of the user which made the chirp. The *third line* is the *chirp* itself. The *fourth line* represents a list of *tags* associated with said chirp. The *fifth line* is a sequence of user ids which *liked* the chirp. The *sixth line* is a sequence of user ids which *disliked* the chirp. In the text file there is always a blank-line between different chirps. Based off this data, the function constructs a `profiles_dictionary` where the orders of the tags and ids representing likes and dislikes, are in the same orders in which they appear in the text file. If there are no tags associated with a chirp, the empty string is associated with it instead. See `chirps.txt`, which is included in the same directory as this notebook, as an example. Once again, you may assume that the given text file is perfect and that there are no formatting errors.

```python
# DONE, PASSES TESTS
def create_chirp_dictionary(file_name):
    f = open(file_name, 'r')
    line = f.readlines()
    new_list = [l.replace('\n','') for l in line]
    new_dict = {}

    i = 0
    while i < len(line)-3:
        x = i + 1
        y = i + 2
        z = i + 3
        a = i + 4
        b = i + 5

        if new_list[z] != '':
            z1 = new_list[z].split(", ")
        else:
            z1 = ['']

        if new_list[a] != '':
            a1 = [new_list[a]]
        else:
            a1 = []

        if new_list[b] != '':
            b1 = [new_list[b]]
        else:
            b1 = []

        new_dict[int(new_list[i])] = ((int(new_list[x]),new_list[y],z1,a1,b1))
        i += 7


    chirp_dictionary = {}
    for key, value in new_dict.items():
        userid, message, tags, likes, dislikes = value
        likes = [int(num) for num in likes[0].split(',')] if likes else []
        dislikes = [int(num) for num in dislikes[0].split(',')] if dislikes else []
        chirp_dictionary[key] = (userid, message, tags, likes, dislikes)


    f.close()
    return(chirp_dictionary)

create_chirp_dictionary("chirps.txt")
```

#### Question 3: `get_top_chirps(profile_dictionary, chirp_dictionary, user_id) -> List[str]`
`get_top_chirps(profile_dictionary, chirp_dictionary, user_id)` returns a list of the most liked chirp for every user `user_id` follows.<br>
The third parameter is a user id which is in the given `profile_dictionary`. Let the user with this user id be denoted by *u*. The function uses the data from the first two parameters to find the chirp with the most likes for each user followed by *u*. The function creates a list of these chirps and returns it. You may assume:
1. For a given user, there does not exist any two chirps with the same number of likes.
2. The order of the chirps in the returned list is arbitrary. As long as all the correct chirps are present and no incorrect chirps are present, the function is correct.
